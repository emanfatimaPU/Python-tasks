```python
#encapsulation:
#Encapsulation is a concept in object-oriented programming (OOP) where you hide the internal details of a class and only expose
#what is necessary to the outside world.in use get and set method.
class student:
    def __init__(self,name):
        self.__name=name
    def get(self):
        return self.__name
    def set(self,name):
        self.__name= name
s1=student('eman')
print(s1.get())
```

```python
#simple inheritance:-
class employee:
    def __init__(self,name,id):
        self.name=name
        self.id=id
    def show_details(self):
        print(f"The name of employee is:",self.name, "And his Id:",self.id)
class programmer(employee):
    def show_language(self):
     print("His language is English.")

e=employee("Numan","345")
e.show_details()
e2=programmer("Ahmar","456")
e2.show_details()
e2.show_language()
```

```python
#simple inheritance:-
class Teacher:
    def __init__(self,name,lectures,pay):
        self.name=name
        self.lectures=lectures
        self.pay=pay
    def show_details(self):
        print(f"Name:",self.name, "And lectuers:",self.lectures,"And pay:",self.pay)
class Student(Teacher):
    def __init__(self,name,lectures,pay,marks):
     super().__init__(name,lectures,pay)
     self.marks=marks
    def show_data(self):
         print(f"good student. His Marks:{self.marks}")
m=Teacher("Mam.Irqa","4 lec.",30000)
m.show_details()
m2=Student("Ahmar","4 lec.",30000, 456)
m2.show_details()
m2.show_data()
```

```python
class vehicle:
    def __init__(self,modal,colour):
        self.modal=modal
        self.colour=colour
    def show_detail(self):
        print("Modal:",self.modal)
        print("Colour of car:",self.colour)
class Car(vehicle):
    def __init__(self, modal, colour,speed):
        super().__init__(modal, colour)
        self.speed=speed
    def show_data(self):
        print("mine fvrt car.")
        print("speed:",self.speed)
C=vehicle("2025",
          "black")
C.show_detail()
C2=Car(2025, "black",250)
C2.show_data()
```
