```python
class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def add(self):
        return self.a + self.b

    def subtract(self):
        return self.a - self.b

    def multiply(self):
        return self.a * self.b

    def divide(self):
        if self.b == 0:
            return "Zero se divide nahi hota"
        return self.a / self.b


x = int(input("Enter a: "))
y = int(input("Enter b: "))

calc = Calculator(x, y)

print("Addition:", calc.add())
print("Subtraction:", calc.subtract())
print("Multiplication:", calc.multiply())
print("Division:", calc.divide())
```

```python
#print table using for loop
class Table:
    def __init__(self,number):
        self.number=number
    def print_table(self):
        for i in range(1,11):
            print(self.number,"x",i,"=",self.number*i)
num=int(input("enter number:"))
num1=Table(num)
num1.print_table()
```

```python
#calculate bill in class 
class Bill:
    def __init__(self,units):
        self.units=units
    def calculate(self):
        if self.units<=100:
            bill= self.units*5
        elif self.units<=200:
            bill=self.units*8
        else:
            bill=self.units*10
        print("Units:",self.units)
        print("Total Bill:",bill)
units=int(input("enter a number:"))
num1=Bill(units)
num1.calculate()
```
