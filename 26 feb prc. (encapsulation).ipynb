```python
class Employe:
    def __init__(self):
        self.__salary='10000'
        self.__bonus='2000'
    def display(self):
        print("salary:",self.__salary)
        print("bonus:",self.__bonus)
        print("total salary:",self.__salary+self.__bonus)
    def get_salary(self):
        return self.__salary
    def set_salary(self,salary):
        self.__salary=salary
    def set_bonus(self,bonus):
        self.__bonus=bonus
n=Employe()
print("1st salary.")
n.display()
print("add bonus.")
n.set_salary(10000)
n.set_bonus(2000)
n.display()
```
