```python
#poliphormism aik class mein zada method aik he name k but perameter diff hoty hn isko overloading static bainding early bending kehty hn.
#isky perametr ka data data type or order chang ho skta hai.
#python m overloading support ni krta. or over ridding ko krta hai.
#over ridding mein name and perameter b same hun gy.inheritance k bgair polimarphism nahi ho gi.

#Ridding 
class Area:
    def __init__(self,a,b):
       self.a=a
       self.b=b
    def area(self):
        return self.a*self.b
 
class triangle(Area):
    def __init__(self, a, b):
        super().__init__(a, b)
    def area(self):
        x=super().area()
        tri=x/2
        return tri
t=triangle(12,2)
print(t.area())
```

```python
#task of overriding:  (task no.1)
class Animal:                                              #(ye base class hai.)
    def sound(self):
        print("Animals makes sound.")
class dog(Animal):                                         #(ye subclass/childclass)
    def sound(self):
        print("dogs bark.")
class cat(Animal):
    def sound(self):
        print("cats meow.")
class cow(Animal):
    def sound(self):
        print("cows moooosss")

D=dog()                                                  #(objects and method)
D.sound()
C=cat()
C.sound()
W=cow()
W.sound()

```

```python
#task no.4 shape area calculation
class shape():
    def area(self):
        print("shpas of area calculation.")
class Rectangle(shape):
    def __init__(self, length, width):
        self.length=length
        self.width=width
    def area(self):
        return self.length * self.width
class Circule(shape):
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        return 3.14 * self.radius **2
class Triangle(shape):
    def __init__(self,base,height):
        self.base=base
        self.height=height
    def area(self):
        return 1/2*self.base**self.height
r=Rectangle(10,30)
print(r.area())
c=Circule(9)
print(c.area())
t=Triangle(40,75)
print(t.area())
        
```

```python
#task no.6 salary calculation system
class Employee:
    def calculate_salary(self):
        pass
class FullTimeEmployee(Employee):
    def __init__(self, fixed_salary):
        self.fixed_salary = fixed_salary
    def calculate_salary(self):
        return self.fixed_salary

class PartTimeEmployee(Employee):
    def __init__(self, hourly_rate, hours):
        self.hourly_rate = hourly_rate
        self.hours = hours
    def calculate_salary(self):
        return self.hourly_rate * self.hours


class Intern(Employee):
    def __init__(self, stipend):
        self.stipend = stipend
    def calculate_salary(self):
        return self.stipend

employees = [
    FullTimeEmployee(10000),
    PartTimeEmployee(800, 13),
    Intern(10000)
]

for emp in employees:
    print("Salary:", emp.calculate_salary())
```
