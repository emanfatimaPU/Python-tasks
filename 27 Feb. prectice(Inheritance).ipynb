```python
class vehicle:
    def __init__(self,modal,colour):
        self.modal=modal
        self.colour=colour
    def show_detail(self):
        print("Modal:",self.modal)
        print("Colour of car:",self.colour)
class Car(vehicle):
    def __init__(self, modal, colour,speed):
        super().__init__(modal, colour)
        self.speed=speed
    def show_data(self):
        print("mine fvrt car.")
        print("speed:",self.speed)
C=vehicle("2025",
          "black")
C.show_detail()
C2=Car(2025, "black",250)
C2.show_data()
```
