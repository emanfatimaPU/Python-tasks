```python
class Animal:
    def __init__(self,name,sound):
        self.name=name # type: ignore
        self.sound=sound
    def display(self):
        print(f"name:{self.name},sound:{self.sound}")
    def bark(self):
        print("meow")

animal=Animal("luccy","meow")
animal.display()
```
